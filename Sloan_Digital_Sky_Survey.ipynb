```python
#Transposing the matrices for implementation purposes

Y_train=np.transpose(np.array(Y_train))
Y_test=np.transpose(np.array(Y_test))
X_train=np.transpose(X_train)
X_test=np.transpose(X_test)
```

```python
#Using relu activation in the input and hidden layers

def relu(x):
    
    return(np.maximum(0,x))
```

```python
#relu gradient to be used in back propagation

def relu_grad(x):
    
    return((np.maximum(x,0)>0).astype(int))
```

```python
#softmax function to be used in output layer

def softmax(x):
    
    e_x=np.exp(x)
    return(e_x/np.sum(e_x,axis=0))
```

```python
def layer_size(X1,y1):
    
    n_x=X1.shape[0]
    n_y=y1.shape[0]
    return(n_x,n_y)
```

```python
#function to initialise weights and biases
#weights are initialised using He normal initializer.Biases are initialsed to zero vectors

def initialise_params(n_x,l1,l2,n_y):
    
    W1=np.random.randn(l1,n_x)*np.sqrt(2/(n_x+l1))#he_normal initializer 
    b1=np.zeros((l1,1))
    W2=np.random.randn(l2,l1)*np.sqrt(2/(l1+l2))#he_normal initializer
    b2=np.zeros((l2,1))
    W3=np.random.randn(n_y,l2)*np.sqrt(2/(l2+n_y))#he_normal initializer
    b3=np.zeros((n_y,1))

    parameters={"W1":W1,"b1":b1,"W2":W2,"b2":b2,"W3":W3,"b3":b3}#A dictionary to store weights and biases
    return(parameters)
```

```python
#Forward propagation

def forward_propagation(X1, parameters):
    
    W1=parameters["W1"]
    b1=parameters["b1"]
    W2=parameters["W2"]
    b2=parameters["b2"]
    W3=parameters["W3"]
    b3=parameters["b3"]

  

    Z1=np.dot(W1,X1)+b1
    A1=relu(Z1)
    Z2=np.dot(W2,A1)+b2
    A2=relu(Z2)
    Z3=np.dot(W3,A2)+b3
    A3=relu(Z3)
    soft=softmax(A3)

    AZ={"A1":A1,"Z1":Z1,"A2":A2,"Z2":Z2,"A3":A3,"Z3":Z3}
    return(soft,AZ)
```

```python
#cost function returns the cost after each iteration of gradient descent.
#L2 regularisation is used with regularisation constant 'lamdb'

def Cost(y,pred,lamdb,parameters):
    
    m = y.shape[1]

    W1 = parameters["W1"]
    W2 = parameters["W2"]
    W3 = parameters["W3"]
    W1 = np.sum(np.square(W1))
    W2 = np.sum(np.square(W2))
    W3 = np.sum(np.square(W3))

    error=-np.sum(np.multiply(np.log(pred),y))/m + lamdb/(2*m)*(W1+W2+W3)
    return(error)
```

```python
#Back propagation
#L2 regularisation is used with regularisation constant 'lamdb'

def backprop(parameters,cache,x1,y1,lamdb):
    
    m=x1.shape[1]

    W1=parameters["W1"]
    W2=parameters["W2"]
    W3=parameters["W3"]

    A1=cache["A1"]
    A2=cache["A2"]
    A3=cache["A3"]
    Z1=cache["Z1"]
    Z2=cache["Z2"]
  
    dZ3 = A3-y1
    dW3 = (1/m)*np.dot(dZ3,A2.T) - lamdb/m*(W3)
    db3 = (1/m)*np.sum(dZ3,axis=1,keepdims=True)
    dZ2 = np.multiply(np.dot(W3.T,dZ3),relu_grad(Z2))
    dW2 = (1/m)*np.dot(dZ2,A1.T) - lamdb/m*(W2)
    db2 = (1/m)*np.sum(dZ2,axis=1,keepdims=True)
    dZ1 = np.multiply(np.dot(W2.T,dZ2),relu_grad(Z1))
    dW1 = (1/m)*np.dot(dZ1,x1.T) - lamdb/m*(W1)
    db1 = (1/m)*np.sum(dZ1,axis=1,keepdims=True)

    grads = {"dW1": dW1,"db1": db1,"dW2": dW2,"db2": db2,"dW3":dW3,"db3":db3}
    return(grads)
```

```python
def update(parameters,grads,alpha):
    
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    W3 = parameters["W3"]
    b3 = parameters["b3"]
    
    dW1 = grads["dW1"]
    db1 = grads["db1"]
    dW2 = grads["dW2"]
    db2 = grads["db2"]
    dW3 = grads["dW3"]
    db3 = grads["db3"]
    
    W1 = W1 - alpha*dW1
    b1 = b1 - alpha*db1
    W2 = W2 - alpha*dW2
    b2 = b2 - alpha*db2
    W3 = W3 - alpha*dW3
    b3 = b3 - alpha*db3

    parameters={"W1":W1,"b1":b1,"W2":W2,"b2":b2,"W3":W3,"b3":b3}

    return parameters
```

```python
#the driver function.

def NN(X1,Y1,l1,l2,iter,lamdb,alpha):

    n_x,n_y = layer_size(X1,Y1)
  
    parameters = initialise_params(n_x,l1,l2,n_y)
    for i in range(0,iter):      
        soft,cache = forward_propagation(X1,parameters)
        cost = Cost(Y1,soft,lamdb,parameters)
        cost_list.append(cost*1000)
        grads = backprop(parameters,cache,X1,Y1,lamdb)
        parameters = update(parameters,grads,alpha)
        print ("Cost after iteration %i: %f" %(i, cost))

    return(parameters)  
```

```python
cost_list=[]
iters= 1300
it = range(iters)
parameters = NN(X_train,Y_train,128,64,iters,0.02,0.07)
plt.plot(it,cost_list)
plt.xlabel('epoch')
plt.ylabel('cost')
```

TRAINING ACCURACY 

```python
#getting the train set predictions

ypred, cache = forward_propagation(X_train,parameters)
```

```python
#converting into label encoding

ypred = np.argmax(ypred,axis=0)
Y_ = np.argmax(Y_train,axis=0)
```

```python
#finding the precision recall and f1-scores

from sklearn.metrics import classification_report
name=["Galaxy","Quasar","Star"]
print(classification_report(Y_,ypred,target_names=name,digits=4))
```

```python
print("Overall train accuracy : ",(Y_==ypred).sum()/75)
```

TEST ACCURACY 


```python
ypred1, cache = forward_propagation(X_test,parameters)
```

```python
#converting into label encoding

ypred1 = np.argmax(ypred1,axis=0)
Y_1 = np.argmax(Y_test,axis=0)
```

```python
#finding precision recall f1-score

name=["Galaxy","Quasar","Star"]
print(classification_report(Y_1,ypred1,target_names=name,digits=4))
```
